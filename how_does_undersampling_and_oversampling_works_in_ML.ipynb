## ðŸ”„Understanding Undersampling and Oversampling
In machine learning, especially classification tasks, we often encounter imbalanced datasets, where one class occurs much more frequently than another.

âœ… Why is this a problem?
ML models tend to bias towards the majority class, which results in poor performance on the minority class.

### ðŸ”§ Solutions:
##### Oversampling â€“ Increase the number of samples in the minority class.
##### Duplicate existing samples (Random Oversampling)
##### Create synthetic samples (e.g., SMOTE)
##### Undersampling â€“ Reduce the number of samples in the majority class.

##### Randomly remove samples from the majority class.

## ðŸ§ª Let's Demo It on a Toy Dataset
We'll use sklearn.datasets.make_classification to create an imbalanced dataset and then apply oversampling and undersampling with visualization using matplotlib.

I'll now generate code to:

#### Create the dataset.
#### Show class imbalance.
#### Apply oversampling and undersampling.
#### Visualize the results.

```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
# Step 1: Create a synthetic imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, 
                           n_redundant=0, n_clusters_per_class=1, 
                           weights=[0.9, 0.1], flip_y=0, random_state=42)

print("Original class distribution:", Counter(y))
```

```python
# Step 2: Plot original data
plt.figure(figsize=(40, 12))

plt.subplot(1, 3, 1)
plt.title("Original Imbalanced Data")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

## Apply Oversampling

```python
# Step 3: Oversampling
ros = RandomOverSampler(random_state=42)
X_over, y_over = ros.fit_resample(X, y)

plt.subplot(1, 3, 2)
plt.title("After Oversampling")
plt.scatter(X_over[:, 0], X_over[:, 1], c=y_over, cmap="coolwarm", alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
print("After oversampling:", Counter(y_over))
```

## Apply Undersampling

```python
# Step 4: Undersampling
rus = RandomUnderSampler(random_state=42)
X_under, y_under = rus.fit_resample(X, y)

plt.subplot(1, 3, 3)
plt.title("After Undersampling")
plt.scatter(X_under[:, 0], X_under[:, 1], c=y_under, cmap="coolwarm", alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
print("After undersampling:", Counter(y_under))

plt.tight_layout()
plt.show()
```

## ðŸ§¾ Summary:
#### Original Dataset: 90% class 0, 10% class 1

#### Oversampling: Balanced by duplicating class 1

#### Undersampling: Balanced by removing from class 0
